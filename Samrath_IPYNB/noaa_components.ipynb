```python
import pandas as pd
import numpy as np

# Read the Excel file
file_path = 'updated_component_17_22_100.xlsx'
df = pd.read_excel(file_path)

# Iterate over each unique component
for component, target_speed in speed_dict.items():
    # Filter rows that have the current component
    component_rows = df[df['Component'] == component]

    if not component_rows.empty:
        # Find the row index where 'Speed (degs/hour)' is closest to the target speed
        closest_index = (component_rows['Speed (degs/hour)'] - target_speed).abs().idxmin()
        
        # Set 'Components' to NaN for all rows except the closest one
        df.loc[(df['Component'] == component) & (df.index != closest_index), 'Component'] = np.nan

# Save the modified DataFrame to a new Excel file
output_file_path = 'updated_component_17_22_100.xlsx'
df.to_excel(output_file_path, index=False)

print(f"Modified Excel file saved to {output_file_path}")
```

```python
import pandas as pd

# Load the reference file
reference_df = pd.read_excel('updated_component_17_17_100.xlsx')
reference_order = reference_df['Component'].dropna().unique()

# Define the list of files to be rearranged
files_to_rearrange = [
    'updated_component_17_18_100.xlsx',
    'updated_component_17_19_100.xlsx',
    'updated_component_17_20_100.xlsx',
    'updated_component_17_21_100.xlsx',
    'updated_component_17_22_100.xlsx'
]

for file in files_to_rearrange:
    # Load the current file
    df = pd.read_excel(file)
    
    # Filter out rows with empty 'Component' values
    df_non_empty = df[df['Component'].notna()]
    
    # Set the 'Component' column as the index for easier reordering
    df_non_empty.set_index('Component', inplace=True)
    
    # Reindex the DataFrame to match the reference order
    df_reordered = df_non_empty.reindex(reference_order)
    
    # Reset the index to restore 'Component' as a column
    df_reordered.reset_index(inplace=True)
    
    # Append rows with empty 'Component' values to the reordered DataFrame
    df_empty = df[df['Component'].isna()]
    final_df = pd.concat([df_reordered, df_empty], ignore_index=True)
    
    # Save the rearranged DataFrame back to an Excel file
    final_df.to_excel(file, index=False)

print("Rearrangement complete.")
```
