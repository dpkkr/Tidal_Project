```python
import pandas as pd
import matplotlib.pyplot as plt

# Number of files and number of plots
num_files = 6
num_plots = 10

# Initialize a list to hold the data from each file
data = []

# Loop through each file and read the 'Amplitude' column
for i in [17,18,19,20,21,22]:
    file_path = f'updated_component_17_{i}_100.xlsx'
    df = pd.read_excel(file_path)
    mylist = df['Amplitude'].tolist()
    if i==17 or i==18:
        mylist = [100*i for i in mylist]
    data.append(mylist)
# Convert the list to a DataFrame for easier manipulation
data_df = pd.DataFrame(data).T

# Plot the data
for i in range(num_plots):
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, num_files + 1), data_df.iloc[i, :], marker='o', linestyle='-') #number of files gives error cuz some files have no values, and are thus eliminated
    plt.title(f'Plot of {i+1}th Component from Each File')
    plt.xlabel('Year number')
    plt.ylabel('Amplitude')
    plt.xticks(range(1, num_files + 1))
    plt.ylim(0)
    plt.grid(True)
    plt.show()
```
