```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,2*np.pi, 100)
y2 = np.sin(x)**2
y3 = np.cos(x)**2
y1 = y2+y3
plt.figure(figsize=(10,6))

plt.plot(x, y1, label='$y=\\sin^2(x)+cos^2(x)$', linestyle='--')

plt.plot(x, y2, label='$y=\\sin^2(x)$', marker = 4, markerfacecolor = "blue", color = "white")

plt.plot(x, y3, label='$y=\\cos^2(x)$', marker = 5,markerfacecolor = "green", color = "white")

plt.title('Plots of $y=\\sin^2(x)+cos^2(x)$, $y=\\sin^2(x)$, and $y=\\cos^2(x)$')
plt.xlabel('x')
plt.ylabel('y')

plt.legend()

plt.grid(True)
plt.show()
```
